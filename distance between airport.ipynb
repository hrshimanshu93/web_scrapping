```python
import requests
from bs4 import BeautifulSoup
import openpyxl

# Load the input Excel file
input_file_path = r'C:\Users\Himanshu\Documents\sector_distance.xlsx'
workbook = openpyxl.load_workbook(input_file_path)
sheet = workbook.active

# Iterate through the rows in the Excel file
for row in sheet.iter_rows(min_row=2, min_col=1, max_col=2, values_only=True):
    origin, destination = row

    # Construct the URL with the origin and destination
    url = f'https://www.airmilescalculator.com/distance/{origin}-to-{destination}'

    # Make a GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the meta tag containing the distance information
        distance_meta_tag = soup.find("meta", {"name": "description"})

        if distance_meta_tag:
            # Extracting the exact distance in miles, kilometers, and nautical miles
            distance_values = distance_meta_tag["content"].split(" is ")[1].split(".")[0]
            miles, kilometers, nautical_miles = [value.strip() for value in distance_values.split("/")]

            # Write the extracted data back to the Excel file
            for row in sheet.iter_rows(min_row=1, max_col=2, max_row=sheet.max_row, values_only=True):
                if row[0] == origin and row[1] == destination:
                    row_number = row[0].row
                    break

            # Write the values to the corresponding columns in the found row
            sheet.cell(row=row_number, column=3, value=miles)
            sheet.cell(row=row_number, column=4, value=kilometers)
            sheet.cell(row=row_number, column=5, value=nautical_miles)
            #sheet.append([miles, kilometers, nautical_miles])
        else:
            # Handle the case when distance information is not found
            print(f"Distance information not found for {origin} to {destination}")
    else:
        # Handle the case when the website is not accessible
        print(f"Failed to fetch data for {origin} to {destination}")

# Save the output to a new Excel file
output_file_path = r'C:\Users\Himanshu\Documents\sector_distance_output1.xlsx'
workbook.save(output_file_path)
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

# Load the input Excel file
input_file_path = r'C:\Users\Himanshu\Documents\sector_distance.xlsx'
workbook = openpyxl.load_workbook(input_file_path)
sheet = workbook.active

# Iterate through the rows in the Excel file
for row in sheet.iter_rows(min_row=2, min_col=1, max_col=2, values_only=True):
    origin, destination = row

    # Construct the URL with the origin and destination
    url = f'https://www.airmilescalculator.com/distance/{origin}-to-{destination}'

    # Make a GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the meta tag containing the distance information
        distance_meta_tag = soup.find("meta", {"name": "description"})

        if distance_meta_tag:
            # Extracting the exact distance in miles, kilometers, and nautical miles
            distance_values = distance_meta_tag["content"].split(" is ")[1].split(".")[0]
            miles, kilometers, nautical_miles = [value.strip() for value in distance_values.split("/")]

            # Find the row number based on the 'origin' and 'destination'
            for r in range(2, sheet.max_row + 1):
                if sheet.cell(row=r, column=1).value == origin and sheet.cell(row=r, column=2).value == destination:
                    row_number = r
                    break

            # Write the values to the corresponding columns in the found row
            sheet.cell(row=row_number, column=3, value=miles)
            sheet.cell(row=row_number, column=4, value=kilometers)
            sheet.cell(row=row_number, column=5, value=nautical_miles)
        else:
            # Handle the case when distance information is not found
            print(f"Distance information not found for {origin} to {destination}")
    else:
        # Handle the case when the website is not accessible
        print(f"Failed to fetch data for {origin} to {destination}")

# Save the output to a new Excel file
output_file_path = r'C:\Users\Himanshu\Documents\sector_distance_output1.xlsx'
workbook.save(output_file_path)
```
